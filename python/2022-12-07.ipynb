##### 대사증후군 판펼
높은 혈압(130/85mmHg 이상)
높은 혈당(공복 혈당 100mg/dL 이상)
높은 중성지방(트리글리세라이드 150mg/dL 이상)
낮은 HDL 콜레스테롤 수치(남성은 40mg/dL 미만, 여성은 50mg/dL 미만)
복부 비만(남성 90cm 이상, 여성 85cm 이상)
판별
0 : 정상, 1~2 : 주의군, 3~5 : 위험군

```python
#columns함수로 열이름 바로 변경
df2.columns = ['성별코드', '허리둘레', '수축기혈압', '이완기혈압', '트리글리세라이드', 'HDL콜레스테롤',
       '공복혈당']
df2.head()
```

```python
#리스트로 변경해서 열이름 변경 및 띄어쓰기 제거
lt = list(df2.columns)
lt
```

```python
type(lt)
```

```python
lt = [c.replace(' ', '') for c in lt]
lt
#c=문자열, 
#띄어쓰기 한번에 처리
```

```python
lt[6] = "공복혈당"
lt
```

```python
#columns함수로 띄어쓰기 제거 및 열이름 바로 변경
df2.columns = [c.replace(' ', '') for c in df2.columns]
df2.columns
```

```python
df2.columns[6] = "공복혈당"
df2.columns
```

```python
df2.columns = lt
df2.head()
```

```python
df2.info()
```

```python
df2 = df2.dropna(axis=0)
#axis=0  -> 결측치가 있는 행 제거
#axis=1  -> 결측치가 있는 열 제거
df2.info()
```

```python
import numpy as np
```

```python
#np.where -> ifelse기능
# 높은 혈압(130/85mmHg 이상)
df2['높은혈압'] = np.where((df2['수축기혈압']>=130) | (df2['이완기혈압']>=85),1,0)
```

```python
df2.head()
```

```python
# 높은 혈당(공복 혈당 100mg/dL 이상)
df2['높은혈당'] = np.where((df2['공복혈당']>=100),1,0)
# 높은 중성지방(트리글리세라이드 150mg/dL 이상)
df2['높은중성지방'] = np.where((df2['트리글리세라이드']>=150),1,0)
# 낮은 HDL 콜레스테롤 수치(남성은 40mg/dL 미만, 여성은 50mg/dL 미만)
df2['낮은HDL콜레스테롤'] = np.where(((df2['성별코드']==1) & (df2['HDL콜레스테롤']<40)) | ((df2['성별코드']==2) & (df2['HDL콜레스테롤']<50)),1,0)
# 복부 비만(남성 90cm 이상, 여성 85cm 이상)
df2['복부비만'] = np.where(((df2['성별코드']==1) & (df2['허리둘레']>=90)) | ((df2['성별코드']==2) & (df2['허리둘레']>=85)),1,0)
```

```python
df2
```

```python
df2['대사증후군'] = df2['복부비만']+df2['낮은HDL콜레스테롤']+df2['높은중성지방']+df2['높은혈당']+df2['높은혈압']
```

```python
df2
```

```python

df2['판별'] = np.where((df2['대사증후군'] == 0),"정상",np.where((df2['대사증후군'] < 3),"주의군", "위험군"))
```

```python
df2.head()
```

```python
df2['성별'] = np.where(df2['성별코드'] == 1,"남","여")
```

```python
pd.DataFrame(df2['성별'], df2.groupby('판별').count()['성별코드'])
```
