```python
import pandas as pd
df=pd.read_csv("../03/03_암발생자수_.csv", encoding="cp949")
```

```python
# 칼럼 이름 수정
df.columns = ['암종별', '성별', '연령별', '2019', '2019.1']
df.columns
```

```python
#열 추출
#R == df22 <- df2 %>% filter(df2$암종별 == "모든 암(C00-C96)" & !(연령별 %in% c("계", "연령미상")))
df22 = df[(df['암종별'] == '모든 암(C00-C96)') & (df['연령별'].isin(['계', '연령미상'])==False)]
# isin = 열이 list의 값들을 포함하고 있는 모든 행들을 골라냄
#df22 = df[(df['암종별'] == '모든 암(C00-C96)') & (df['연령별'] != '계') & (df['연령별'] != '연령미상')]
df22
```

```python
import numpy as np
```

```python
df22['연령대'] = np.where(df22['연령별'].isin(["0-4세","5-9세","10-14세","15-19세","20-24세","25-29세","30-34세","35-39세"]),"30대이하",
                np.where(df22['연령별'].isin(["40-44세","45-49세","50-54세","55-59세"]),"40대~50대",
                np.where(df22['연령별'].isin(["60-64세","65-69세","70-74세","75-79세"]),"60대~70","80대이상")))
df22
```

```python
#R == df22$y2019 <- as.numeric(df22$y2019)
df22['2019'] = df22['2019'].astype('int')
df22.info()
```

```python
df22g = df22.groupby(['연령대','성별']).sum()[['2019']]
#df22.groupby(['연령대','성별']).sum()['2019']
df22g
```

```python
#컬럼명 인덱스가 아닌 행 번호 인덱스로 사용하고 싶은 경우에는 reset_index를 사용해 컬럼명으로 지정한 인덱스를 제거
df22g = df22g.reset_index()
df22g
```
